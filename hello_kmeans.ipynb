```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.metrics import calinski_harabasz_score
```

```python
x,y=datasets.make_blobs(n_samples=1000,n_features=2,cluster_std=[0.4,0.1,0.1,0.1],centers=[[-1, -1], [0, 0], [1, 1], [2, 2]],random_state=1)
plt.scatter(x[:,0],x[:,1])
plt.show()
```

```python
y_pre=KMeans(n_clusters=5,random_state=2).fit_predict(x)
plt.scatter(x[:,0],x[:,1],c=y_pre)
plt.show()
```

```python
calinski_harabasz_score(x,y_pre)
```
